```python
# Which property types have best reviews, check property type and scores rating column
airbnb[['property_type','review_scores_rating']]
```

```python
# Find out what property type the dataset has
property_type=airbnb.property_type.unique()
property_type
```

```python
# creat function to work out the mean rating for different property types

mean_list=[]
count_list=[]
for p_property_type in property_type:
    mean= airbnb[airbnb['property_type']==p_property_type]['review_scores_rating'].mean()
    mean_list.append(mean)
    
    count=len(airbnb[airbnb['property_type']==p_property_type]['review_scores_rating'])
    count_list.append(count) 
print(mean_list, count_list)
```

```python
property_type_reviews=pd.DataFrame({'Property Type': property_type, 'Count':count_list, 'Mean Rating':mean_list})
property_type_reviews
```

```python
# Put best rating on top
property_type_reviews.sort_values(by='Mean Rating', ascending= False, inplace=True)
property_type_reviews
 
```

```python
property_type_reviews.plot.barh(x='Property Type', y='Mean Rating')
plt.xlabel('Rating',fontsize=16)
plt.ylabel("")
plt.title('Property Type Reviews Rating', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
property_type_reviews.plot.barh(x='Property Type', y='Count')
plt.xlabel('Count',fontsize=16)
plt.ylabel("")
plt.title('Most Popular Property Type', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# set Property Type as index
property_type_reviews.set_index('Property Type',inplace=True)
property_type_reviews
```
